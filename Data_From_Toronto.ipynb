## First import pandas, numpy and requests.

## Next I set the link to the wikipedia page as wiki_url to pass it to the panda.read_html function.  


```python
import pandas as pd
import requests
import numpy as np
wiki_url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

## Next I used df[0] to turn it into a dataframe


```python
df=pd.read_html(wiki_url)
df=df[0]
df.head()
```

## I changed the "Not assigned" cells to the numpy NaN placeholder so I could use the panda.dropna method.

## The index no longer started at zero so I reset the index and dropped the old one.

```python
df.replace("Not assigned", np.nan, inplace = True)
df.dropna(axis = 0, inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```
